```python
# -*- coding: utf-8 -*-
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
zhfont1 = mpl.font_manager.FontProperties(fname='C:\\Windows\\Fonts\\simsun.ttc')
```

```python
df_sha = pd.read_csv('sha/processed_data.csv')
df_fen = pd.read_csv('fen/processed_data.csv')
cols = list(df_sha.columns)
t_sha = [5*i/60 for i in range(df_sha.shape[0])]
t_fen = [5*i/60 for i in range(df_fen.shape[0])]
```

```python
plt.figure(figsize=(16,9),dpi=300)
for nrow in range(3):
    for ncol in range(2):
        plot_num = 2*nrow+ncol+1
        plt.subplot(3,2,plot_num)
        plt.plot(t_sha,df_sha[cols[plot_num]])
        plt.gca().invert_yaxis()
        plt.title(u'珊瑚砂-' + cols[plot_num], fontproperties=zhfont1)
```

```python
plt.figure(figsize=(16,9),dpi=300)
for nrow in range(3):
    for ncol in range(2):
        plot_num = 2*nrow+ncol+1
        plt.subplot(3,2,plot_num)
        plt.plot(t_fen,df_fen[cols[plot_num]])
        plt.gca().invert_yaxis()
        plt.title(u'珊瑚粉-' + cols[plot_num], fontproperties=zhfont1)
```
