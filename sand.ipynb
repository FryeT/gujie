```python
# -*- coding: utf-8 -*-
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
#zhfont1 = mpl.font_manager.FontProperties(fname='C:\\Windows\\Fonts\\simsun.ttc',size=12)
```

```python
import numpy as np
```

```python
df_sha = pd.read_csv('sha/sha_processed_data.csv')
#df_fen = pd.read_csv('fen/fen_processed_data.csv')
cols = list(df_sha.columns)
t_sha = np.array([5*i/60 for i in range(df_sha.shape[0])])
#t_fen = np.array([5*i/60 for i in range(df_fen.shape[0])])
```

```python
df_sha['t_h'] = t_sha/60
df_sha['t_m'] = t_sha
df_sha
```

```python
# df_fen['t_h'] = t_fen/60
# df_fen['t_m'] = t_fen
# df_fen
```

```python
#sand_s_200
fig, ax = plt.subplots(figsize=(8,4), dpi=300)
ax.plot(df_sha['t_m'], df_sha[cols[0]])
ax.invert_yaxis()
ax.set_xlabel('time / min', fontsize=12)
ax.set_ylabel('settlement / mm', fontsize=12)
fig.tight_layout()
ax.set_title('sand-{0}'.format(cols[0]), fontsize=14)
# 插入
inset_ax = fig.add_axes([0.5, 0.5, 0.35, 0.35]) # X, Y, 宽度，高度
inset_ax.plot(df_sha['t_m'], df_sha[cols[0]])
inset_ax.set_title('zoom near origin')
# 设置坐标轴范围
inset_ax.set_xlim(0, 10)
inset_ax.set_ylim(0.5, 0)
inset_ax.set_xlabel('time / min',fontsize=12)
inset_ax.set_ylabel('settlement / mm', fontsize=12)
fig.savefig('figs/sand-{0}'.format(cols[0])+'.svg')
```
